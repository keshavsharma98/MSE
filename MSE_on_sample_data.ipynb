```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error
```

```python
#generate n evenly spaced values from the range 0 to 2pie
n=200
x=np.linspace(0,2*np.pi,n)

#taking sin values of those x 
sine = np.sin(x)

#plotting x vs sin(x) values
plt.plot(x,sine)
```

```python
#adding noise to sin wave
noise=0.6
noisy_sine = sine + np.random.uniform(-noise,noise,n)

#plotting both original and noisy graph together
plt.plot(x,noisy_sine,color='r',linewidth=1)
plt.plot(x,sine,linewidth=3)
```

```python
#calculating MSE using numpy
error = (1/n)*sum(np.power(noisy_sine - sine,2))
error
```

```python
#claculating MSE using sklearn library function
error = mean_squared_error(sine,noisy_sine)
error
```
